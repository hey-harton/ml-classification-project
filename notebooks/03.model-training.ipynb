```python
import os
import pandas as pd
import joblib
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

# Registri Dataset
---

```python
# dataset ecoli processed
path_train_ec = r'../datasets/engineered/ecoli_train_engineered.csv'

# dataset glass processed
path_train_gl = r'../datasets/engineered/glass_train_engineered.csv'

# dataset thyroid processed
path_train_th = r'../datasets/engineered/thyroid_train_engineered.csv'

models_dir = os.path.join('..', 'models')
os.makedirs(models_dir, exist_ok=True)
```

# Training Model `ecoli`
---

## Memuat Dataset

```python
df_train_ec = pd.read_csv(path_train_ec)

print("[1] Data Ecoli berhasil dimuat.")
print(f"    Dimensi Ecoli: {df_train_ec.shape}")
```

## Penanganan Pemisahan Data

```python
X_train_ec = df_train_ec.drop(columns=['label_encoded'])
y_train_ec = df_train_ec['label_encoded']

print("[2] Pemisahan fitur dan target Ecoli selesai.")
print(f"    Dimensi Fitur (X_train) : {X_train_ec.shape}")
print(f"    Dimensi Target (y_train): {y_train_ec.shape}")
```

## Penanganan Inisiasi Model & Hyper Paramter

```python
models_ec = {
    'Naive_Bayes': GaussianNB(
        priors=None,            # Probabilitas awal tiap kelas (None = dihitung dari data)
        var_smoothing=1e-9      # Ditambahkan ke varians untuk stabilitas perhitungan (bisa diperbesar jika overfit)
    ),
    
    'KNN': KNeighborsClassifier(
        n_neighbors=3,          # Jumlah tetangga terdekat
        weights='uniform',      # 'uniform' (semua tetangga bobotnya sama) atau 'distance' (bobot berdasarkan jarak)
        algorithm='auto',       # Algoritma pencarian ('auto', 'ball_tree', 'kd_tree', 'brute')
        p=2,                    # p=1 untuk Manhattan distance, p=2 untuk Euclidean distance
        metric='minkowski'      # Metrik jarak standar (kombinasi dari parameter p di atas)
    ),
    
    'Decision_Tree': DecisionTreeClassifier(
        criterion='gini',       # Kriteria pemisahan cabang ('gini', 'entropy', atau 'log_loss')
        splitter='best',        # Strategi pemisahan ('best' atau 'random')
        max_depth=5,            # Batas kedalaman maksimal pohon (mencegah overfitting)
        min_samples_split=2,    # Jumlah sampel minimum untuk memecah sebuah node
        min_samples_leaf=1,     # Jumlah sampel minimum yang harus ada di daun (node ujung)
        class_weight=None,      # Bisa diubah menjadi 'balanced' jika data sangat tidak seimbang dan tidak pakai SMOTE
        random_state=42         # Mengunci seed acak agar hasil tetap konsisten saat di-run ulang
    )
}

print("[3] Algoritma Naive Bayes, KNN, dan Decision Tree berhasil diinisialisasi dengan parameter lengkap.")
```

## Pelatihan Model

```python
print("=== Memulai Pelatihan Model untuk Dataset: ECOLI ===")


trained_models_ec = {}

for model_name, model_obj in models_ec.items():
    model_obj.fit(X_train_ec, y_train_ec)
    trained_models_ec[model_name] = model_obj
    print(f" Model {model_name} berhasil dilatih.")

print("\n[4] Seluruh model untuk Ecoli selesai dilatih!")
```

## Penanganan Eksport PKL

```python
print("=== Memulai Ekspor Model Ecoli (Format PKL) ===")

models_dir = os.path.join('..', 'models')
os.makedirs(models_dir, exist_ok=True)

for model_name, trained_model in trained_models_ec.items():
    model_filename = f"ecoli_{model_name.lower()}.pkl"
    model_path = os.path.join(models_dir, model_filename)
    joblib.dump(trained_model, model_path)
    print(f" Model {model_name} berhasil diekspor di: {model_path}")

print("\n[5] Seluruh model untuk Ecoli berhasil diekspor ke format PKL!")
```
