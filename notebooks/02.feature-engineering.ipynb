```python
import os
import pandas as pd
from sklearn.preprocessing import StandardScaler
from imblearn.combine import SMOTETomek
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
```

# Registri Dataset

```python
# dataset ecoli processed
path_train_ec = r'../dataset/processed/ecoli_train_processed.csv'
path_test_ec = r'../dataset/processed/ecoli_test_processed.csv'

# dataset glass processed
path_train_gl = r'../dataset/processed/glass_train_processed.csv'
path_test_gl = r'../dataset/processed/glass_test_processed.csv'

# dataset thyroid processed
path_train_th = r'../dataset/processed/thyroid_train_processed.csv'
path_test_th = r'../dataset/processed/thyroid_test_processed.csv'
```

# Feature Engineering `Ecoli` Dataset
---

## Memuat Dataset

```python
df_train = pd.read_csv(path_train_ec)
df_test = pd.read_csv(path_test_ec)

print("[1] Data Ecoli berhasil dimuat.")
print(f"    Dimensi Train : {df_train.shape}")
print(f"    Dimensi Test  : {df_test.shape}")
```

## Penanganan Pemisahan Data

```python
X_train = df_train.drop(columns=['label_encoded'])
y_train = df_train['label_encoded']

X_test = df_test.drop(columns=['label_encoded'])
y_test = df_test['label_encoded']

print("[2] Pemisahan fitur (X) dan target (y) selesai.")
```

## Penanganan Scaling

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)

print("[3] Feature Scaling (StandardScaler) selesai diterapkan.")
```

## Penanganan Smote Data

```python
smote = SMOTE(k_neighbors=1, random_state=42)

X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled_df, y_train)

print("[4] Penyeimbangan data dengan SMOTE selesai.")
print("\nDistribusi kelas SEBELUM SMOTE:\n", y_train.value_counts().sort_index())
print("\nDistribusi kelas SESUDAH SMOTE:\n", y_train_balanced.value_counts().sort_index())
```

## Penanganan Eksport

```python
df_train_engineered = X_train_balanced.copy()
df_train_engineered['label_encoded'] = y_train_balanced

df_test_engineered = X_test_scaled_df.copy()
df_test_engineered['label_encoded'] = y_test.values

output_dir = os.path.join('..', 'dataset', 'engineered')
os.makedirs(output_dir, exist_ok=True)

train_out_path = os.path.join(output_dir, 'ecoli_train_engineered.csv')
test_out_path = os.path.join(output_dir, 'ecoli_test_engineered.csv')

df_train_engineered.to_csv(train_out_path, index=False)
df_test_engineered.to_csv(test_out_path, index=False)

print("\n[5] Proses Ekspor Selesai!")
print(f"    Data latih (engineered) disimpan ke: {train_out_path}")
print(f"    Data uji (engineered) disimpan ke  : {test_out_path}")
```

# Feature Engineering `glass` Dataset
---

## Memuat Dataset

```python
df_train_gl = pd.read_csv(path_train_gl)
df_test_gl = pd.read_csv(path_test_gl)

print("[1] Data Glass Processed berhasil dimuat.")
print(f"    Dimensi Train : {df_train_gl.shape}")
print(f"    Dimensi Test  : {df_test_gl.shape}")
```

## Penanganan Pemisahan data

```python
X_train_gl = df_train_gl.drop(columns=['label_encoded'])
y_train_gl = df_train_gl['label_encoded']

X_test_gl = df_test_gl.drop(columns=['label_encoded'])
y_test_gl = df_test_gl['label_encoded']

print("[2] Pemisahan fitur dan target Glass selesai.")
```

## Penanganan Scaling Data

```python
scaler_gl = StandardScaler()

X_train_gl_scaled = scaler_gl.fit_transform(X_train_gl)
X_test_gl_scaled = scaler_gl.transform(X_test_gl)

X_train_gl_scaled_df = pd.DataFrame(X_train_gl_scaled, columns=X_train_gl.columns)
X_test_gl_scaled_df = pd.DataFrame(X_test_gl_scaled, columns=X_test_gl.columns)

print("[3] Feature Scaling (StandardScaler) untuk Glass selesai.")
```

## Penanganan Smote Data

```python
smote_internal_gl = SMOTE(k_neighbors=1, random_state=42)

hybrid_sampler_gl = SMOTETomek(smote=smote_internal_gl, random_state=42)

X_train_gl_balanced, y_train_gl_balanced = hybrid_sampler_gl.fit_resample(X_train_gl_scaled_df, y_train_gl)

print("[4] Hybrid Resampling (SMOTE-Tomek) pada Glass selesai.")
print("\nDistribusi kelas SEBELUM Resampling:\n", y_train_gl.value_counts().sort_index())
print("\nDistribusi kelas SESUDAH Resampling:\n", pd.Series(y_train_gl_balanced).value_counts().sort_index())
```

## Penanganan Eksport

```python
df_train_gl_engineered = X_train_gl_balanced.copy()
df_train_gl_engineered['label_encoded'] = y_train_gl_balanced

df_test_gl_engineered = X_test_gl_scaled_df.copy()
df_test_gl_engineered['label_encoded'] = y_test_gl.values

output_dir = os.path.join('..', 'dataset', 'engineered')
os.makedirs(output_dir, exist_ok=True)

train_out_path_gl = os.path.join(output_dir, 'glass_train_engineered.csv')
test_out_path_gl = os.path.join(output_dir, 'glass_test_engineered.csv')

df_train_gl_engineered.to_csv(train_out_path_gl, index=False)
df_test_gl_engineered.to_csv(test_out_path_gl, index=False)

print("\n[5] Proses Ekspor Glass Selesai!")
print(f"    Data latih disimpan di : {train_out_path_gl}")
print(f"    Data uji disimpan di   : {test_out_path_gl}")
```
