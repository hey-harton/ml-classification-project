```python
import os
import pandas as pd
import joblib
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

## Registry Model 


```python
# Path Folder
data_dir = os.path.join('..', 'dataset', 'engineered')
models_dir = os.path.join('..', 'models')

# Path Data Uji (Test Set)
path_test_ec = os.path.join(data_dir, 'ecoli_test_engineered.csv')
path_test_gl = os.path.join(data_dir, 'glass_test_engineered.csv')
path_test_th = os.path.join(data_dir, 'thyroid_test_engineered.csv')
```

## Automasi Evaluasi

```python
def evaluate_models(dataset_name, test_data_path, model_names):
    print(f"=== Evaluasi Model untuk Dataset: {dataset_name.upper()} ===")
    
    df_test = pd.read_csv(test_data_path)
    X_test = df_test.drop(columns=['label_encoded'])
    y_test = df_test['label_encoded']
    
    results = []
    
    for model_name in model_names:
        model_filename = f"{dataset_name}_{model_name.lower()}.pkl"
        model_path = os.path.join(models_dir, model_filename)
        
        model = joblib.load(model_path)
        
        y_pred = model.predict(X_test)
        
        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, average='weighted', zero_division=0)
        rec = recall_score(y_test, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
        
        results.append({
            'Dataset': dataset_name.capitalize(),
            'Model': model_name,
            'Accuracy': round(acc, 4),
            'Precision': round(prec, 4),
            'Recall': round(rec, 4),
            'F1-Score': round(f1, 4)
        })
        
    df_results = pd.DataFrame(results)
    return df_results

print("[2] Fungsi evaluasi berhasil dibuat.")
```

## Evaluasi Model `ecoli`

```python
algorithms = ['Naive_Bayes', 'KNN', 'Decision_Tree']

df_eval_ec = evaluate_models('ecoli', path_test_ec, algorithms)

print("\n--- Hasil Evaluasi ECOLI ---")
display(df_eval_ec)
```

## Evaluasi Model `glass`

```python
df_eval_gl = evaluate_models('glass', path_test_gl, algorithms)

print("\n--- Hasil Evaluasi GLASS ---")
display(df_eval_gl)
```

# Evauasi Model `thyroid`

```python
df_eval_th = evaluate_models('thyroid', path_test_th, algorithms)

print("\n--- Hasil Evaluasi THYROID ---")
display(df_eval_th)
```

## Penggabungan Evaluasi Model

```python
df_all_results = pd.concat([df_eval_ec, df_eval_gl, df_eval_th], ignore_index=True)

reports_dir = os.path.join('..', 'reports')
os.makedirs(reports_dir, exist_ok=True)
df_all_results.to_csv(os.path.join(reports_dir, 'all_models_evaluation.csv'), index=False)

print("\n=== RINGKASAN EVALUASI SELURUH MODEL ===")
print("File ringkasan juga telah disimpan di: reports/all_models_evaluation.csv\n")
display(df_all_results)
```
