```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

# Registri Dataset

```python
path_1 = r'https://raw.githubusercontent.com/hey-harton/ml-classification-project/refs/heads/main/dataset/raw/ecoli.csv'
path_2 = r'https://raw.githubusercontent.com/hey-harton/ml-classification-project/refs/heads/main/dataset/raw/glass.csv'
path_3 = r'https://raw.githubusercontent.com/hey-harton/ml-classification-project/refs/heads/main/dataset/raw/thyroid.csv'

# dimuat dalam dataframe
df_ecoli = pd.read_csv(path_1)
df_2 = pd.read_csv(path_2)
df_3 = pd.read_csv(path_3)
```

# Preprocessing `Ecoli` Dataset

## Memuat dataset

```python
df_ecoli.head()
```

## Penanganan Nilai Kosong

```python
if df_ecoli.isnull().sum().any():
    df_ecoli = df_ecoli.dropna()
    print("[2] Terdapat missing values. Baris kosong telah dihapus.")
```

## Penanganan Nilai Duplikasi

```python
jumlah_duplikat = df_ecoli.duplicated().sum()
if jumlah_duplikat > 0:
    df_ecoli = df_ecoli.drop_duplicates()
    print(f"[2] Ditemukan {jumlah_duplikat} baris duplikat. Duplikat telah dihapus.")

print(f"    Dimensi data setelah cleansing: {df_ecoli.shape}")
```

## Pemisahan Fitur

```python
X = df_ecoli.drop(columns=['label'])
y = df_ecoli['label']
```

## Penangana Transformasi Encoding

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("Mapping Kelas (Label Asli -> Angka):")
for index, label in enumerate(label_encoder.classes_):
    print(f"{label} -> {index}")
```

## Pembagian Data

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print("\n[4] Data Splitting selesai.")
print(f"    Dimensi X_train : {X_train.shape}")
print(f"    Dimensi X_test  : {X_test.shape}")
```

## Penanganan Eksport

```python
df_train_processed = X_train.copy()
df_train_processed['label_encoded'] = y_train

df_test_processed = X_test.copy()
df_test_processed['label_encoded'] = y_test

train_filename = 'ecoli_train_processed.csv'
test_filename = 'ecoli_test_processed.csv'

df_train_processed.to_csv(train_filename, index=False)
df_test_processed.to_csv(test_filename, index=False)

print("\n[5] Proses Ekspor Selesai!")
print(f"    Data latih tersimpan sebagai : {train_filename}")
print(f"    Data uji tersimpan sebagai   : {test_filename}")
```
