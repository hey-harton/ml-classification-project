```python
import os
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

# Registri Dataset
---

```python
path_1 = r'../dataset/raw/ecoli.csv'
path_2 = r'../dataset/raw/glass.csv'
path_3 = r'../dataset/raw/thyroid.csv'
```

# Preprocessing `Ecoli` Dataset
---

## Memuat dataset

```python
df_ecoli = pd.read_csv(path_1)

print("[1] Data Glass berhasil dimuat.")
print(f"    Dimensi Glass: {df_ecoli.shape}")
```

## Penanganan Nilai Kosong

```python
if df_ecoli.isnull().sum().any():
    df_ecoli = df_ecoli.dropna()
    print("[2] Terdapat missing values. Baris kosong telah dihapus.")
```

## Penanganan Duplikasi Nilai

```python
jumlah_duplikat = df_ecoli.duplicated().sum()
if jumlah_duplikat > 0:
    df_ecoli = df_ecoli.drop_duplicates()
    print(f"[2] Ditemukan {jumlah_duplikat} baris duplikat. Duplikat telah dihapus.")

print(f"    Dimensi data setelah cleansing: {df_ecoli.shape}")
```

## Penanganan Pemisahan Fitur

```python
X = df_ecoli.drop(columns=['label'])
y = df_ecoli['label']
```

## Penanganan Transformasi Encoding

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("Mapping Kelas (Label Asli -> Angka):")
for index, label in enumerate(label_encoder.classes_):
    print(f"{label} = {index}")
```

## Penanganan Pembagian Data

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print("\n[4] Data Splitting selesai.")
print(f"    Dimensi X_train : {X_train.shape}")
print(f"    Dimensi X_test  : {X_test.shape}")
```

## Penanganan Eksport

```python
df_train_processed = X_train.copy()
df_train_processed['label_encoded'] = y_train

df_test_processed = X_test.copy()
df_test_processed['label_encoded'] = y_test

output_dir = os.path.join('..', 'dataset', 'processed')
os.makedirs(output_dir, exist_ok=True)

train_filepath = os.path.join(output_dir, 'ecoli_train_processed.csv')
test_filepath = os.path.join(output_dir, 'ecoli_test_processed.csv')

df_train_processed.to_csv(train_filepath, index=False)
df_test_processed.to_csv(test_filepath, index=False)

print("\n[5] Proses Ekspor Selesai!")
print(f"    Data latih tersimpan di : {train_filepath}")
print(f"    Data uji tersimpan di   : {test_filepath}")
```

# Preprosesing `glass` Dataset
---

## Memuat dataset

```python
df_glass = pd.read_csv(path_2)

print("[2] Data Glass berhasil dimuat.")
print(f"    Dimensi Glass: {df_glass.shape}")
```

```python
if df_glass.isnull().sum().any():
    df_glass = df_glass.dropna()
    print("[2] Terdapat missing values. Baris kosong telah dihapus.")
```
